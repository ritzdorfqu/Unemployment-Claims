```python
county_corona = pd.read_csv('https://query.data.world/s/bvonzvikz6ceyemxg6fwux4funslgh')
```

```python
county_corona.head()
```

```python
march_start = '03/01/2020'
```

```python
march_end = '03/31/2020'
```

```python
missouri_county_corona = county_corona[
    county_corona['state'] == 'Missouri'
]
```

```python
missouri_county_corona.head()
```

```python
march_mo_cases = missouri_county_corona[
    missouri_county_corona['date'] >= march_start
].sort_values(
    'date'
)
```

```python
total_march_mo_cases = march_mo_cases[
    march_mo_cases['date'] <= march_end
].sort_values(
    'date'
)
```

```python
total_march_mo_cases.head()
```

```python
total_march_mo_cases[
    total_march_mo_cases['location_name'] == 'Boone'
].sort_values('date')
```

```python
march_missouri_county_cases = total_march_mo_cases[
    total_march_mo_cases['date'] == march_end
].sort_values(
    'location_name'
)
```

```python
march_missouri_county_cases.head()
```

```python
counties = alt.topo_feature(data.us_10m.url, 'counties')
source = march_missouri_county_cases

alt.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    from_=alt.LookupData(source, 'cumulative_cases_per_100_000')
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```
