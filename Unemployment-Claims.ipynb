## What is the national percentage change of new claims per week since the virus started? Graph it

```python
virus_date = '2020-03-07'
```

```python
national_claims[
    national_claims['date'] >= virus_date
].groupby('date').pct_lastweek.sum().reset_index()
```

## What is the national percentage change from last year for the weeks of the virus? Graph it

```python
national_claims[
    national_claims['date'] >= virus_date
].groupby('date').pct_lastyear.sum().reset_index()
```

## What is the national total number of people on claims?

8,177,965 people. 2020-04-04 isn't updated yet and has just 0.0

```python
national_claims[
    national_claims['date'] == '2020-03-28'
].groupby('date').cont_claims.sum().reset_index()
```

## Is that the most since 2000?

Yes by more than 1.5 million claims

```python
national_claims.groupby('date').cont_claims.sum().reset_index().sort_values('cont_claims', ascending=False)
```
